# :-Year with best Ratings

```python
# Load the IMDb India Movies dataset
movies_df = pd.read_csv('/content/drive/MyDrive/IMDb Movies India/IMDb Movies India.csv', encoding='ISO-8859-1')

# Group by year and calculate the average rating for each year
average_ratings_by_year = movies_df.groupby('Year')['Rating'].mean()

# Find the year with the highest average rating
best_year = average_ratings_by_year.idxmax()
best_rating = average_ratings_by_year.max()

print(f"The year with the best rating is {best_year} with an average rating of {best_rating:.2f}")

```

## :> Does length of movie have any impact with the rating?


To determine if the length of a movie has any impact on its rating, we can perform a more in-depth analysis by exploring the relationship between the two variables. Here are some steps we can take using Python and pandas:
If the scatter plot shows a clear trend or pattern, it indicates a potential relationship.The correlation coefficient ranges from -1 to 1. A positive value suggests a positive correlation (as one variable increases, the other tends to increase), and a negative value suggests a negative correlation.A positive value close to 1 suggests a positive correlation (as movie length increases, the rating tends to increase).
A negative value close to -1 suggests a negative correlation (as movie length increases, the rating tends to decrease).
A value close to 0 suggests a weak or no linear correlation.


```python
# Load the IMDb India Movies dataset
movies_df = pd.read_csv('/content/drive/MyDrive/IMDb Movies India/IMDb Movies India.csv', encoding='ISO-8859-1')

# Explore the relationship between Length and Rating
sns.scatterplot(x='Duration', y='Rating', data=movies_df)
plt.title('Relationship Between Movie Length and Rating')
plt.xlabel('Duration (minutes)')
plt.ylabel('Rating')
plt.show()
import numpy as np

# Assuming you have two arrays 'duration' and 'rating'
duration = [1, 2, 3, 4, 5]
rating = [3, 4, 2, 5, 1]

# Calculate the correlation coefficient
correlation_coefficient = np.corrcoef(duration, rating)[0, 1]

# Print the correlation coefficient
print(f"Correlation Coefficient between Duration and Rating: {correlation_coefficient:.2f}")


```

## :> Top 10 movies according to rating per year and overall.

```python
# Convert the "Rating" column to numeric
movies_df['Rating'] = pd.to_numeric(movies_df['Rating'], errors='coerce')

# Find the top 10 movies overall
top_10_overall = movies_df.sort_values(by='Rating', ascending=False).head(10)

# Find the top 10 movies per year
top_10_per_year = movies_df.groupby('Year').apply(lambda group: group.nlargest(10, 'Rating'))

# Print the results
print("Top 10 Movies Overall:")
print(top_10_overall[['Name', 'Rating']])

print("\nTop 10 Movies Per Year:")
print(top_10_per_year[['Year', 'Name', 'Rating']])
```

## Top ten high average rated genre

```python
# plot top ten high average rated genre
genre_rate = df.groupby('Genre').agg({'Rating':'mean'}).sort_values(by='Rating', ascending=False)\
                                .reset_index().head(10)

ax = sns.barplot(data=genre_rate, x='Rating', y='Genre', palette='gist_gray')
ax.bar_label(ax.containers[0])
plt.title('Top ten high average rated genre')
plt.show()
```

# Number of popular movies released each year.

```python
# Filter out rows with missing or NaN values in the "Rating" or "Year" columns
movies_df = movies_df.dropna(subset=['Rating', 'Year'])

# Define a threshold for considering movies as popular (you can adjust this threshold)
popularity_threshold = 7.0

# Count the number of popular movies released each year
popular_movies_per_year = movies_df[movies_df['Rating'] >= popularity_threshold].groupby('Year')['Name'].count()

# Print the results
print("Number of Popular Movies Released Each Year:")
print(popular_movies_per_year)
```

```python
# Number of movie release per year
year_count = df.groupby('Year').agg({'Name':'count'}).rename(columns={'Name':'count'}).\
                                        sort_values(by='count', ascending=False).reset_index()
plt.figure(figsize=(20,8))
ax = sns.barplot(data=year_count, x='Year', y='count')
ax.bar_label(ax.containers[0])
plt.title('Number of movie releases per year')
plt.xticks(rotation=90)
plt.show()
```

## Counting the number of votes which movies preformed better in rating per year and overall.

```python
# Find the maximum rating for each year
max_rating_per_year = movies_df.groupby('Year')['Rating'].max()

# Extract the movie names with the maximum rating for each year
top_movies_per_year = movies_df[movies_df.set_index(['Year', 'Rating']).index.isin(max_rating_per_year.items())]

# Find the top movie overall based on maximum rating
top_movie_overall = movies_df.loc[movies_df['Rating'].idxmax()]

# Print the results
print("Top Movie Overall:")
print(top_movie_overall[['Name', 'Year', 'Rating', 'Votes']])

print("\nTop Movie Per Year:")
print(top_movies_per_year[['Name', 'Year', 'Rating', 'Votes']])
```

# Average Rating Per Year

```python
# Average rating per year
year_rating = df.groupby('Year').agg({'Rating': 'mean'}).reset_index().sort_values(by='Rating',
                                                                     ascending=False,
                                                                     )
year_rating['Rating'] = round(year_rating['Rating'],1)

plt.figure(figsize=(25,9))
ax = sns.barplot(data=year_rating, x='Year', y='Rating')
ax.bar_label(ax.containers[0])
plt.title('Average rating per year')
plt.xticks(rotation=90)
plt.show()
```

# Which director directed the most movies

```python
# Group by director and count the number of movies directed by each
directors_movies_count = df['Director'].value_counts()

# Get the director with the most movies
most_prolific_director = directors_movies_count.idxmax()
number_of_movies_directed = directors_movies_count.max()

print(f"The director who directed the most movies is {most_prolific_director} with {number_of_movies_directed} movies.")
```
